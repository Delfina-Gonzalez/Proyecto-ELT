# Proyecto ELT - AviationStack

Este notebook ejecuta las etapas de extracción y carga de datos de la API AviationStack, usando **Python + Pandas + pyarrow**, sin Spark.  

Se guardan los datos crudos y procesados en formato Parquet, simulando un Data Lake.

---



```python
## 1️⃣ Configuración inicial

import os
import subprocess

# Definir paths
project_dir = "C:/Users/Fede/Desktop/Proyecto ELT"
extract_script = os.path.join(project_dir, "src", "extract.py")
load_script = os.path.join(project_dir, "src", "load.py")

# Directorios de salida
flights_raw = os.path.join(project_dir, "data_lake/flights/flights_raw.parquet")
airports_raw = os.path.join(project_dir, "data_lake/airports/airports_raw.parquet")
```

```python
# Ejecutar extract.py
subprocess.run(["python", extract_script], check=True)
```

```python
import pandas as pd

df_flights = pd.read_parquet(flights_raw, engine="pyarrow")
df_airports = pd.read_parquet(airports_raw, engine="pyarrow")

# Primeras filas
df_flights.head(5), df_airports.head(5)
# muestra vuelos recientes con columnas como flight_date, flight_status, departure, arrival, airline.

# información de aeropuertos como iata_code, city_iata_code, country_name.
```

```python
# Ejecutar load.py
subprocess.run(["python", load_script], check=True)
```

```python
processed_flights_dir = os.path.join(project_dir, "data_lake/flights/processed")
processed_airports_file = os.path.join(project_dir, "data_lake/airports/processed/data.parquet")

# Leer vuelos de una fecha específica
flight_files = [os.path.join(processed_flights_dir, d, "data.parquet") 
                for d in os.listdir(processed_flights_dir)
                if os.path.isdir(os.path.join(processed_flights_dir, d))]

df_flights_proc = pd.concat([pd.read_parquet(f, engine="pyarrow") for f in flight_files])
df_airports_proc = pd.read_parquet(processed_airports_file, engine="pyarrow")

df_flights_proc.head(5), df_airports_proc.head(5)
```
